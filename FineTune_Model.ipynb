```python
import pandas as pd
import nltk
from nltk.tokenize import sent_tokenize
from tqdm import tqdm
nltk.download('punkt')
```

```python
from transformers import AutoTokenizer, pipeline, AutoModelForSeq2SeqLM
```

```python
from datasets import load_dataset, load_from_disk, load_metric
```

```python
model = 'google/pegasus-cnn_dailymail'
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
model_pegasus = AutoModelForSeq2SeqLM.from_pretrained(model).to('cpu')
```

```python
dataset = load_dataset('samsum')
```

```python
dataset['train']
```

```python
def convert_examples_to_features(example_batch):
    input_encodings = tokenizer(example_batch['dialogue'] , max_length = 1024, truncation = True )

    with tokenizer.as_target_tokenizer():
        target_encodings = tokenizer(example_batch['summary'], max_length = 128, truncation = True )

    return {
        'input_ids' : input_encodings['input_ids'],
        'attention_mask': input_encodings['attention_mask'],
        'labels': target_encodings['input_ids']
    }
```

```python
dataset_en = dataset.map(convert_examples_to_features, batched = True)
```

```python
dataset_en['train']['attention_mask'][0]
```

```python
from transformers import TrainingArguments, Trainer
```

```python
from transformers import DataCollatorForSeq2Seq

seq2seqdata_collator = DataCollatorForSeq2Seq(tokenizer, model =model)
```

```python
training_args = TrainingArguments(output_dir="pegasus-samsum")
```

```python
dataset_new = dataset_en['test'][:400]
```

```python
trainer = Trainer(model = model_pegasus,args = training_args, tokenizer = tokenizer, data_collator = seq2seqdata_collator,
                  train_dataset= dataset_new , eval_dataset= dataset_en['validation'])
```

```python
trainer.train()
```
