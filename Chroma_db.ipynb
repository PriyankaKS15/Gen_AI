```python
from google.colab import userdata
HUGGINGFACE_API_KEY = userdata.get('HUGGINGFACE_API_KEY')
```

```python
import os
os.environ['HUGGINGFACE_API_KEY'] = HUGGINGFACE_API_KEY
```

```python
from langchain import HuggingFaceHub
```

```python
from langchain.vectorstores import Chroma
# from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.embeddings import HuggingFaceEmbeddings
```

```python
from langchain.document_loaders import DirectoryLoader
from langchain.document_loaders import TextLoader
```

```python
loader = DirectoryLoader('new_articles', glob='**/*.txt', loader_cls=TextLoader)
document = loader.load()
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
text_splitter =  RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
```

```python
texts = text_splitter.split_documents(document)
```

```python
texts[3]
```

```python
persist_directory = './db'
embedding = HuggingFaceEmbeddings()

vectordb = Chroma.from_documents(texts, embedding, persist_directory=persist_directory)
```

```python
vectordb.persist()
vectordb = None
```

```python
vectordb = Chroma(persist_directory=persist_directory, embedding_function=embedding)
```

```python
retriever = vectordb.as_retriever()
```

```python
docs = retriever.get_relevant_documents("what is the comment of Yedoye Travis on AI")
```

```python
from langchain.chains import RetrievalQA
```

```python
qachain = RetrievalQA.from_chain_type(llm = HuggingFaceHub(
    repo_id="mistralai/Mistral-7B-Instruct-v0.2",
    huggingfacehub_api_token=HUGGINGFACE_API_KEY,
),
                                      chain_type="stuff",
                                      retriever = retriever,
                                      return_source_documents =True)
```

```python
llm_response = qachain("what is the comment of Yedoye Travis on AI")
```

```python
def process_llm_reponse(llm_res):
  print(llm_res['result'])
  print('\n\nSources:')
  for source in llm_res['source_documents']:
    print(source.metadata['source'])
```

```python
query  ="what is spacex"
llm_response = qachain(query)
process_llm_reponse(llm_response)
```
